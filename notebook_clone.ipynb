```python
import shutil
import os
import nbformat as nbf

def clone_notebook(new_name_notebook):
    # Define the paths
    template_path = "notebook_template.ipynb"  # Replace with the path to your template notebook
    new_folder_name = new_name_notebook
    new_notebook_name = new_name_notebook + ".ipynb"

    # Create a new directory with the specified name
    os.makedirs(new_folder_name, exist_ok=True)

    # Copy the template notebook to the new directory
    shutil.copy(template_path, os.path.join(new_folder_name, new_notebook_name))

    # Create a new Jupyter notebook object
    nb = nbf.v4.new_notebook()

    # Create a code cell that assigns the variable solution_name
    code_cell = nbf.v4.new_code_cell(f'solution_name = "{new_name_notebook}"')
    
    # Add the code cell to the notebook
    nb.cells.append(code_cell)

    # Save the notebook to the new directory
    with open(os.path.join(new_folder_name, new_notebook_name), 'w') as f:
        nbf.write(nb, f)

if __name__ == "__main__":
    new_name_notebook = "new_notebook"  # Change this to your desired notebook name
    clone_notebook(new_name_notebook)
```
